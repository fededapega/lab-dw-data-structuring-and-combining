```python
df_master['gender'] = df_master['gender'].replace(['female','Femal'],'F')
df_master['gender'] = df_master['gender'].replace(['Male'],'M')
df_master['state'] = df_master['state'].replace(['AZ'],'Arizona')
df_master['state'] = df_master['state'].replace(['Cali'],'California')
df_master['state'] = df_master['state'].replace(['WA'],'Washington')
df_master['education'] = df_master['education'].replace(['Bachelors'],'Bachelor')
df_master['education'] = df_master['education'].replace(['Doctor'],'Doctorate')
df_master['customer_lifetime_value'] = df_master['customer_lifetime_value'].str.replace('%','')
df_master['vehicle_class'] = df_master['vehicle_class'].replace(['Sports Car', 'Luxury SUV', 'Luxury Car'],'Luxury')
df_master
```

```python
df_master['gender'] = df_master['gender'].fillna('unknown')
```

```python
df_master['customer_lifetime_value'] = df_master['customer_lifetime_value'].fillna(0)
```

```python
df_master['customer_lifetime_value'] = df_master['customer_lifetime_value'].astype(float)
```

```python
df_master['number_of_open_complaints'] = df_master['number_of_open_complaints'].fillna(0)
```

```python
df_master['number_of_open_complaints'] = df_master['number_of_open_complaints'].astype(int)
```

```python
#turn null values in income to 0
df_master['income'] = df_master['income'].fillna(0)
```

```python
#turn null values in STATE to unknown
df_master['state'] = df_master['state'].fillna('unknown')
df_master['customer'] = df_master['customer'].fillna('unknown')
df_master['education'] = df_master['education'].fillna('unknown')
df_master['monthly_premium_auto'] = df_master['monthly_premium_auto'].fillna(0)
df_master['policy_type'] = df_master['policy_type'].fillna('unknown')
df_master['total_claim_amount'] = df_master['total_claim_amount'].fillna(0)
```

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*
